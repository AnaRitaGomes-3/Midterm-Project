```python
import pandas as pd
from hts.hierarchy import HierarchyTree
```

```python
# import the data files
data = pd.read_excel('./data.xlsx',sheet_name='Quadro', skiprows=11, usecols="A:D", header=None, index_col=0)
```

```python
first_row = data.loc["Anos",:]
```

```python
data.columns=first_row
```

```python
data.reset_index(inplace=True)
data.columns.values[0]="date"
data.drop(data.index[[0]],inplace=True)
```

```python
new_index=pd.to_datetime(data.date, format='%Y%b')
data.index=new_index
```

```python
data.drop(data.columns[0], axis=1,inplace=True)
data
```

```python
type(data.index)
```

```python
hier = {'Total': ['Norte_Total', 'Centro_Total','Área Metropolitana de Lisboa_Total', 'Alentejo_Total','Algarve_Total','Região Autónoma dos Açores_Total','Região Autónoma da Madeira_Total'],
        'Norte_Total': ['Norte_Alto Minho','Norte_Cávado','Norte_Ave','Norte_Área Metropolitana do Porto','Norte_Alto Tâmega','Norte_Tâmega e Sousa','Norte_Douro','Norte_Terras de Trás-os-Montes'],
        'Centro_Total': ['Centro_Oeste','Centro_Região de Aveiro','Centro_Região de Coimbra','Centro_Região de Leiria','Centro_Viseu Dão Lafões','Centro_Beira Baixa','Centro_Médio Tejo','Centro_Beiras e Serra da Estrela'],
        'Área Metropolitana de Lisboa_Total': ['Área Metropolitana de Lisboa_Área Metropolitana de Lisboa'],
        'Alentejo_Total': ['Alentejo_Alentejo Litoral','Alentejo_Baixo Alentejo','Alentejo_Lezíria do Tejo','Alentejo_Alto Alentejo','Alentejo_Alentejo Central'],
        'Algarve_Total': ['Algarve_Algarve'],
        'Região Autónoma dos Açores_Total': ['Região Autónoma dos Açores_Ilha de Santa Maria','Região Autónoma dos Açores_Ilha de São Miguel','Região Autónoma dos Açores_Ilha Terceira','Região Autónoma dos Açores_Ilha Graciosa','Região Autónoma dos Açores_Ilha de São Jorge','Região Autónoma dos Açores_Ilha do Pico','Região Autónoma dos Açores_Ilha do Faial','Região Autónoma dos Açores_Ilha das Flores','Região Autónoma dos Açores_Ilha do Corvo'],
        'Região Autónoma da Madeira_Total': ['Região Autónoma da Madeira_Ilha da Madeira']
    }
```

```python
tree = HierarchyTree.from_nodes(hier, data, root='Total')
```

```python
print(tree)
```
